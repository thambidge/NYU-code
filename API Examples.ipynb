## Reading xml files 

```python
import xml.etree.ElementTree as et
from lxml import objectify

path = 'test.xml'
xml = objectify.parse(open(path))
```

```python
print(xml)
```

This is how the xml file looks like:
"""
<xml_root>

    <object>
        <id>1</id>
        <name>First</name>
    </object>

    <object>
        <id>2</id>
        <name>Second</name>
    </object>

    <object>
        <id>3</id>
        <name>Third</name>
    </object>

    <object>
        <id>4</id>
        <name>Fourth</name>
    </object>

</xml_root>
"""

```python
# Get the root node
root = xml.getroot()
print(root)
```

```python
# Now we can access child nodes
root.getchildren()[0].getchildren()
```

```python
# Generate empty dataframe
df = pd.DataFrame(columns=('id', 'name'))
df.head()
```

```python
# Now grab info from xml file and append to dataframe
for i in range(0,4):
    obj = root.getchildren()[i].getchildren()
    row = dict(zip(['id', 'name'], [obj[0].text, obj[1].text]))
    row_s = pd.Series(row)
    row_s.name = i
    df = df.append(row_s)
```
